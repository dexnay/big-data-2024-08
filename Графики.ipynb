```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np

%matplotlib inline
```

```python
plt.figure()

plt.plot([0.5, 1, 1.5], [0, 2, 1])
plt.show()
```

```python
plt.figure()

plt.scatter([0.5, 1, 1.5], [0, 2, 1])
plt.show()
```

```python
# (0, 1) (1, 2) (2, 1) (1, 0)

plt.figure()

plt.plot([0, 1, 2, 1, 0], [1, 2, 1, 0, 1])
plt.show()
```

```python
# sin(x) + cos(x)  [0, 2*pi]
x = np.linspace(0, 2*np.pi, 100)
y = np.sin(x) + np.cos(x)

plt.figure()

plt.plot(x, y)
plt.show()
```

```python
plt.figure(figsize=(4,4))

t = np.linspace(0, 2*np.pi, 100)
x = np.cos(t)
y = np.sin(t)


plt.plot(x, y)
plt.show()
```

```python
# 1. 
# 2. 
# 3. 

plt.figure(figsize=(4,4))

t = np.linspace(0, 2*np.pi, 1000)
r = np.cos(6*t)
x = r * np.cos(t)
y = r * np.sin(t)

plt.plot(x, y)
plt.show()
```

```python
plt.figure(figsize=(4,4))

t = np.linspace(0, 2*np.pi, 1000)
r = np.cos(6*t)

x = 2* r * np.cos(t) - r * np.cos(2*t )
y = 2* r * np.sin(t) - r * np.sin(2*t)

plt.plot(x, y)
plt.show()
```

```python
ax = plt.subplot(projection='polar')

phi = np.linspace(0, 10*np.pi, 1000)
r = 1+2*phi
plt.plot(phi, r)
```

```python
x = np.linspace(0, 2, 10)

plt.figure()
plt.plot(x, x, 'r', marker="+", markersize=10, label=r'$x$')
plt.plot(x, x**2, 'b--', label=r'$x^2$')
plt.plot(x, x**3, 'gs', label=r'$x^3$')
plt.grid()
plt.xlabel(r'$x$')
plt.ylabel(r'$y$')
#plt.title()
plt.legend()
plt.show()
```

```python
x = np.linspace(0, 10, 10)

plt.figure()
plt.plot(x, np.exp(x), 'r')
plt.grid()
#plt.yscale('log')
#plt.xscale('log')
plt.show()
```

```python
x = np.linspace(0, 10, 10)
y = np.sin(x)
z = np.cos(x)
fig = plt.figure()

ax = Axes3D(fig)
ax.plot(x, y, z)
plt.show()
```
