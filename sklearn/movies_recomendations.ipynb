````python
# сколько оценок было у каждого пользователя
# сколько оценок было у каждого фильма
user_votes = ratings.groupby('userId')['rating'].agg('count')
movie_votes = ratings.groupby('movieId')['rating'].agg('count')
````

````python
movie_votes.head(20)
````

````python
movie_filter =  movie_votes[movie_votes > 10].index
user_filter = user_votes[user_votes > 50].index

user_item_matrix = user_item_matrix.loc[movie_filter,:]
user_item_matrix = user_item_matrix.loc[:,user_filter]
````
